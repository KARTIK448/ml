```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import datasets
from sklearn.decomposition import PCA

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

# Get the cluster labels and centroids
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# Use PCA to reduce the dataset to 2D for visualization
pca = PCA(2)
X_pca = pca.fit_transform(X)

# Plot the data points and cluster centers
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, s=50, cmap='viridis')

# Plot the cluster centers (after PCA transformation)
centers_pca = pca.transform(centers)
plt.scatter(centers_pca[:, 0], centers_pca[:, 1], c='red', s=200, alpha=0.75, marker='x')

plt.title("KMeans Clustering on Iris Dataset")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```
